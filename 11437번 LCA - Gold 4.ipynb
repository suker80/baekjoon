# 11437번 LCA - Gold 4



```python
import sys
sys.setrecursionlimit(10000)
n = int(input())
input = sys.stdin.readline

rank = {}
parent= {}
graph = [[] for _ in range(n+1)]
for i in range(n-1):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

visit = [False] * (n+1)

def dfs(node,p=0,r=0):

    visit[node] = True
    parent[node] = p
    if not graph[node]:
        rank[node] = r
        return

    for i in graph[node]:
        if visit[i] == False:
            dfs(i, node , r+1)

    rank[node] = r
dfs(1)
m = int(input())

def lca(a,b):
    rank_a = rank[a]
    rank_b = rank[b]

    if rank_a > rank_b:
        for i in range(rank_a - rank_b):

            a = parent[a]
    elif rank_a  < rank_b:
        for i in range(rank_b - rank_a):
            b = parent[b]
    while a != b:
        a = parent[a]
        b = parent[b]


    return a



for i in range(m):
    a,b= map(int,input().split())
    print(lca(a,b))
```
