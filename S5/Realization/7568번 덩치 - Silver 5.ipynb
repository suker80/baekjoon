# 7568번 덩치 - Silver 5

#### 알고리즘 분류: 브루트포스 , 구현






```python
n = int(input())

p = [tuple(map(int,input().split())) for i in range(n) ]

result = []

for i in range(n):
    count = 1
    for j in range(n):
        if p[i][0] < p[j][0] and p[i][1] < p[j][1]:
            count +=1
    result.append(count)

print(*result, sep='\n')
```
