# 17171번 집합의 표현 - Gold 4


```python
n,m = map(int,input().split())
parent = {}
def find(n):
    if parent[n] != n:
        parent[n] = find(parent[n])
    return parent[n]

def union(r1,r2):
    if r1> r2:
        parent[r2] = r1
    else:
        parent[r1] = r2
for i in range(m):
    op,a,b = map(int,input().split())
    
    if op == 0:
        union(a,b)
    elif op==1:
        if find(a) == find(b):
            print("YES")
        else:
            print('NO')
```
