```python
import sys


n = int(input())
m= int(input())


city = [[float('inf')] * (n+1) for _ in range(n+1)]

for i in range(m):
    a,b,c = map(int,sys.stdin.readlined())
    city[a][b] = min(city[a][b],c)
    
print(city)
```
