# 2206번 벽 부수고 이동하기 - Gold 4



```python
n,m=map(int,input().split())
import copy
from collections import deque
graph = [list(map(int,input().split())) for _ in range(n)]



direction = [(1,0),(-1,0),(0,1),(0,-1)]

flag = True 

start = (0,0)
count_map = [[0] * m for _ in range(n)]

min_dist = float('inf')
def bfs(start,flag,graph):
    
    
    queue = deque([start])
    
    while queue:
        
        y,x = queue.popleft()
        
        for dy,dx in direction:
            ny,nx = dy+y,dx+x 
            
            if 0<=ny<n and 0<=nx<m and flag = True :
                
                flag = False
                copy_graph= copy.deepcopy(graph)
                copy_graph[ny][nx] = graph[y][x] + 1 
                
                bfs((ny,nx),flag,copy_graph)

                
            if 0<=ny<n and 0<=nx<m and graph[ny][nx] == 0:
                queue.append((ny,nx))
                graph[ny][nx] = graph[y][x] + 1
                
                if ny == n-1 and nx == m- 1:
                    min_dist = min(min_dist,graph[ny][nx])
print(min_dist) if min_dist != float('inf') else print(-1)
```
