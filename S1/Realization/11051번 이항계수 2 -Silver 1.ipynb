#### 알고리즘 분류 : 다이나믹 프로그래밍 , 조합론



```python
import sys
n, k  = map(int,input().split())
if n== 1:
    print(1)
    sys.exit()
arr = [[0]*(n+1) for _ in range(n+1)]
arr[1][1] = 1
arr[2][1:3] = [2,1]

for i in range(3,n+1):
    arr[i][1] = i
    for j in range(2,i+1):

        arr[i][j] = arr[i-1][j-1] + arr[i-1][j]
print(arr[n][k] % 10007)
```
