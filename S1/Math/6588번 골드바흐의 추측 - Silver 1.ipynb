```python
t = int(input())
import sys
import bisect
```

```python
def sieve(even):
    sieves = [True]* (even+1)
    max_sieve = int(even**0.5)
    for i in range(2,max_sieve):
        if sieves[i] == True:           
            for j in range(i+i,even+1,i):
                sieves[j] = False
    return sieves
    
    

prime_list = sieve(1000000)
len(prime_list)
```
