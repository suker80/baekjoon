# 11403번 경로찾기

#### 알고리즘 분류: 그래프 이해 , 플로이드 와샬



```python
n= int(input())

graph = [list(map(int,input().split())) for _ in range(n)]

result = [[0] * n for _ in range(n)]

for k in range(n):
    for i in range(n):
        for j in range(n):
            if graph[i][k] and graph[k][j] :
                graph[i][j]=1

for i in range(n):
    print(*graph[i])
```
