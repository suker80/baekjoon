# 1927번 최소 힙 - Silver 1


#### 알고리즘 분류: 자료구조 , 우선 순위 큐



```python
import heapq
import sys

n = int(input())

for _ in range(n):
    
    op = int(sys.stdin.readline())
    heap = []
    heapq
    if op:
        heapq.heappush(heap,op)
    else:
        if len(heap):
            print(heapq.heappop(heap))
        else:
            print(0)
```
