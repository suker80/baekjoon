```python
import heapq
import sys

n = int(input())

for _ in range(n):
    
    op = int(sys.stdin.readline())
    heap = []
    heapq
    if op:
        heapq.heappush(heap,op)
    else:
        if len(heap):
            print(heapq.heappop(heap))
        else:
            print(0)
```
