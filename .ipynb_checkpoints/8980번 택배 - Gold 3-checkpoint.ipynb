```python
n,c = map(int,input().split())

m = int(input())

data = []
for _ in range(m):

    data.append(list(map(int,input().split())))


data = sorted(data,key=lambda x:[x[1],x[0]])
truck = 0
current = 0
lift = [c] * (n+1)
total = 0


for d in data:
    start , end ,val = d

    can_lift = min(lift[start:end])

    if can_lift >= val:
        total += val
        for i in range(start,end):
            lift[i] -= val

    elif can_lift > 0 and can_lift < val:
        total += can_lift
        for i in range(start,end):
            lift[i] -= can_lift
    else:
        continue










print(total)
```
