# 12015번 가장 긴 증가하는 부분 수열 2 - Gold 2


#### 알고리즘 분류 : 다이나믹 프로그래밍
#### 시간복잡도 : O(NlogN)



```python
n = int(input())
import bisect
arr = list(map(int,input().split()))

dp = [arr[0]]
l = 1
for i in range(1,n):
    if dp[-1] < arr[i]:
        dp.append(arr[i])
        l+=1
    elif dp[-1]> arr[i]:
        idx = bisect.bisect_left(dp,arr[i])
        dp[idx] = arr[i]
print(l)
```
