```python
n,m,k = map(int,input().split())
import copy
ground = [list(map(int,input().split())) for _ in range(n)]
temp_ground = copy.deepcopy(ground)
tree = [list(map(int,input().split())) for _ in range(m)]

direction = [-1,0,1]

dead_tree =[]
def spring():
    tree.sort(key= lambda x : x[2])
    
    for i,x,y,z in enumerate(tree):
        
        if temp_ground[y-1][x-1] >= z:
            temp_ground[y-1][x-1] -= z
            tree[i][2] += 1
        else:
            dead_tree.append(tree.pop(i))
def summer():
    for i,x,y,z in enumerate(dead_tree)
        temp_ground[y-1][x-1] += z//2
    dead_tree = []
    
def fall():
    
    for x,y,z in tree:
        if z%5==0 :
            for dy in direction:
                for dx in direction:
                    ny,nx = y+dy, x+dx
                    if ny == y and nx == x: continue
                    if 0<=ny<n and 0<=nx<n:
                        tree.append((nx,ny,1))
def winter():
    for i in range(n):
        for j in range(n):
            temp_ground[i][j] += ground[i][j]
for i in range(k):
    sprint()
    summer()
    fall()
    winter()
print(len(tree))
```
