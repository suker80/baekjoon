#### 알고리즘 분류: 그리디 알고리즘 



```python
n = int(input())
import sys
input = sys.stdin.readline
homework = [list(map(int,input().split())) for _ in range(n)]


homework.sort(key=lambda x:(-x[1] ,x[0]))
answer= 0
plan = [0] * 1001
ans = 0
for h in homework:

    day,weight = h
    for j in range(day,0,-1):
        if plan[j] == 0:
            plan[j] = True
            ans += weight
            break
print(ans)
```
