```python
n = int(input())


def prime_list(n, first):
    sieve = [True] * n
    m = int(n ** 0.5)
    for i in range(2, m + 1):
        if sieve[i] == True:
            for j in range(i + i, n, i):
                sieve[j] = False

    return [i for i in range(first, n) if sieve[i] == True]


sieve = prime_list(n+1, 2)

s = 0
left, right = 0, 0
sums = 1
answer= 0
while right >=left  and right < len(sieve):

    if sieve[right] + s <=n:
        s += sieve[right]
        right += 1
    else :
        s -= sieve[left]
        left += 1
    if s== n:
        answer += 1
print(answer)
```
